#### ปัญหา/คำถามที่สนใจมีความสำคัญอย่างไร?
1. ปัจจัยใดมีอิทธิพลต่อการเข้าชิงรางวัลออสการ์ประเภทภาพยนต์ยอดเยี่ยมแห่งปีมากที่สุด
ทำให้เข้าใจปัจจัยหลักที่กำหนดโอกาสในการเข้าชิงรางวัลและการตัดสินใจของคณะกรรมการ รวมถึงช่วยให้ผู้สร้างภาพยนตร์สามารถนำข้อมูลเหล่านี้ไปใช้วางกลยุทธ์การผลิตเพื่อเพิ่มโอกาสในการได้รับการเสนอชื่อ

2. ประเภทของภาพยนตร์ที่ได้รับความนิยมสูง มีแนวโน้มที่จะได้รับคะแนนเฉลี่ยสูงกว่าประเภทอื่นหรือไม่
ชี้ให้เห็นความสัมพันธ์ระหว่าง แนวภาพยนตร์ (Genre) กับ การตอบรับจากผู้ชม เพื่อทำความเข้าใจว่าประเภทใดของหนังได้รับความนิยมมากที่สุดและอาจมีแนวโน้มเข้าสู่รางวัลได้มากกว่า
3. การที่ภาพยนต์ได้รับรางวัลออสการ์ จะส่งผลให้คะแนนเฉลี่ยจากผู้ชมเพิ่มขึ้นภายหลังการประกาศรางวัลหรือไม่
ทำให้เข้าใจอิทธิพลของรางวัลต่อพฤติกรรมและการประเมินของผู้ชม ว่าการได้รับรางวัลสามารถสร้าง “อิทธิพลทางจิตวิทยา” ให้ผู้ชมมองหนังเรื่องนั้นในแง่บวกมากขึ้นหรือไม่
4. ผู้ใช้มี bias ในการให้ rating ของภาพยนต์หรือไม่
ชี้ให้เห็นความเอนเอียงของผู้ชมในการให้คะแนน เช่น การให้คะแนนสูงกับหนังทุกเรื่องที่ดู แม่จะเป็นเรื่องที่ไม่ชอบ ซึ่งส่งผลต่อความน่าเชื่อถือของข้อมูลและการวิเคราะห์เชิงสถิติในโมเดลพยากรณ์

5. หนังที่ได้คะแนน rating มากจะยิ่งมีโอกาสเข้าชิงรางวัลออสการ์มากขึ้นหรือไม่
ชี้ให้เห็นความสัมพันธ์ระหว่าง การประเมินจากผู้ชม กับ การยอมรับจากสถาบันรางวัล ว่ามีแนวโน้มไปในทิศทางเดียวกันหรือไม่ ซึ่งสะท้อนว่าความนิยมของผู้ชมมีผลต่อการตัดสินของคณะกรรมการหรือไม่
6. หนังที่ได้จำนวนรีวิวมาก จะยิ่งมีโอกาสเข้าชิงรางวัลออสการ์มากขึ้นหรือไม่
ชี้ให้เห็นความสัมพันธ์ระหว่าง ความนิยมของผู้ชมในเชิงปริมาณ กับ ความสำเร็จทางรางวัล ว่าการได้รับความสนใจจากสาธารณะมีผลต่อการได้รับการยอมรับจากเวทีออสการ์หรือไม่

## Data Preparation

````python
merged_32m = pd.merge(ratings_32m, movies_32m, on='movieId', how='left')
merged_32m.columns = merged_32m.columns.str.lower()
merged_32m['datetime'] = pd.to_datetime(merged_32m['timestamp'], unit='s')
merged_32m.drop('timestamp', axis=1, inplace=True)
display(merged_32m.head())
````

````python
merged_32m['year'] = merged_32m['title'].str.extract(r'\((\d{4})\)')
merged_32m['year'] = pd.to_numeric(merged_32m['year'], errors='coerce')
merged_32m_filtered = merged_32m[merged_32m['year'] >= 2010].copy()
merged_32m_filtered['year'] = merged_32m_filtered['year'].astype('Int64')
display(merged_32m_filtered.head())
````

````python
merged_32m_filtered.shape
````

````python
oscar_renamed = oscar.rename(columns={'Oscar-winning films': 'title', 'Oscar-nominated films': 'title_nominated'})

# Create a flag for Oscar-winning films
oscar_winners = oscar_renamed[['Year', 'title']].copy()
oscar_winners['is_oscar_winner'] = 1
oscar_winners.rename(columns={'title': 'title_winner'}, inplace=True) # Rename for clarity before melting

# Create a flag for Oscar-nominated films
oscar_nominated = oscar_renamed[['Year', 'title_nominated']].copy()
oscar_nominated['is_oscar_nominated'] = 1
oscar_nominated.rename(columns={'title_nominated': 'title'}, inplace=True) # Rename back to 'title' for merging

oscar_flags = pd.concat([
    oscar_nominated[['title', 'is_oscar_nominated']],
    oscar_winners[['title_winner', 'is_oscar_winner']].rename(columns={'title_winner': 'title'})
]).groupby('title').max().reset_index()

oscar_flags['is_oscar_nominated'] = oscar_flags['is_oscar_nominated'].fillna(0)
oscar_flags['is_oscar_winner'] = oscar_flags['is_oscar_winner'].fillna(0)

print("Prepared Oscar flags dataframe:")
display(oscar_flags.head(10))

merged_32m_with_oscar = pd.merge(merged_32m_filtered, oscar_flags, on='title', how='left')
merged_32m_with_oscar['is_oscar_winner'].fillna(0, inplace=True)
merged_32m_with_oscar['is_oscar_nominated'].fillna(0, inplace=True)

print("\nMerged dataframe with Oscar flags:")
display(merged_32m_with_oscar.head(10))
print("\nInfo of merged_32m_with_oscar:")
````

#### ผู้ใช้มี bias ในการให้ rating ของภาพยนต์หรือไม่

**คำถาม**

ผู้ใช้มี bias ในการให้ rating ของภาพยนต์หรือไม่

**ความสำคัญของคำถาม**

ผู้ใช้มี bias ในการให้ rating ของภาพยนต์หรือไม่ ชี้ให้เห็นความเอนเอียงของผู้ชมในการให้คะแนน เช่น การให้คะแนนสูงกับหนังทุกเรื่องที่ดู แม่จะเป็นเรื่องที่ไม่ชอบ ซึ่งส่งผลต่อความน่าเชื่อถือของข้อมูลและการวิเคราะห์เชิงสถิติในโมเดลพยากรณ์

**ข้อคาดการณ์**

ผู้ใช้ให้คะแนนภาพยนตร์ในระดับเดียวกันทุกเรื่อง ไม่ว่าจะชื่นชอบหรือไม่ ส่งผลให้เกิด bias ในการให้คะแนน rating ภาพยนตร์


````python
# สรุปสถิติของการให้คะแนนต่อผู้ใช้
user_bias = movie_data.groupby('userid')['rating'].agg(['mean', 'std', 'count']).reset_index()
user_bias.columns = ['userid', 'avg_rating', 'std_rating', 'review_count']
user_bias['std_rating'] = user_bias['std_rating'].fillna(0)

# ตรวจสอบสัดส่วนของ 1 ดาว และ 5 ดาว จาก original movie_data
bias_ratios = movie_data.groupby('userid')['rating'].apply(lambda x: pd.Series({
    'ratio_1star': (x == 1).sum() / len(x),
    'ratio_5star': (x == 5).sum() / len(x)
})).reset_index()

bias_ratios_pivoted = bias_ratios.pivot(index='userid', columns='level_1', values='rating').reset_index()
bias_ratios_pivoted.columns.name = None # Remove the columns name after pivot

user_bias = user_bias.merge(bias_ratios_pivoted, on='userid', how='left')

# ตั้งเงื่อนไข Bias (>= 80%)
bias_threshold = 0.8
user_bias['is_1star_bias'] = user_bias['ratio_1star'] >= bias_threshold
user_bias['is_5star_bias'] = user_bias['ratio_5star'] >= bias_threshold

# แสดงผลผู้ใช้ที่มีแนวโน้ม bias
biased_1star = user_bias[user_bias['is_1star_bias'] & (user_bias['review_count'] > 3)]
biased_5star = user_bias[user_bias['is_5star_bias'] & (user_bias['review_count'] > 3)]

print("จำนวนผู้ใช้ที่มีแนวโน้ม 1-star bias:", len(biased_1star))
print("จำนวนผู้ใช้ที่มีแนวโน้ม 5-star bias:", len(biased_5star))
print("\nตัวอย่างผู้ใช้ที่อาจมี 1-star bias:")
display(biased_1star.sort_values(by='review_count', ascending=False).head(10))
print("\nตัวอย่างผู้ใช้ที่อาจมี 5-star bias:")
display(biased_5star.sort_values(by='review_count', ascending=False).head(10))

# การกระจายคะแนนเฉลี่ยและความแปรปรวนของแต่ละผู้ใช้
plt.figure(figsize=(10, 6))
sns.histplot(user_bias['avg_rating'], bins=30, kde=True)
plt.title("Distribution of Average Rating per User")
plt.xlabel("Average Rating Given by User")
plt.ylabel("Number of Users")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(user_bias['std_rating'], bins=30, kde=True)
plt.title("Distribution of Rating Variability (Standard Deviation) per User")
plt.xlabel("Standard Deviation of Ratings per User")
plt.ylabel("Number of Users")
plt.show()

# Visualization: สัดส่วนผู้ใช้ที่มี Bias
bias_summary = {
    '1-star bias': user_bias['is_1star_bias'].sum() if 'is_1star_bias' in user_bias.columns else 0,
    '5-star bias': user_bias['is_5star_bias'].sum() if 'is_5star_bias' in user_bias.columns else 0,
    'No bias': len(user_bias) - (user_bias['is_1star_bias'].sum() + user_bias['is_5star_bias'].sum()) if ('is_1star_bias' in user_bias.columns and 'is_5star_bias' in user_bias.columns) else len(user_bias)
}

plt.figure(figsize=(7, 6))
plt.pie(bias_summary.values(), labels=bias_summary.keys(), autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Users by Bias Type")
plt.show()

print("จำนวนผู้ใช้ทั้งหมด:", len(user_bias))
print(f"ผู้ใช้ที่มีแนวโน้ม 1-star bias: {user_bias['is_1star_bias'].sum()} "
      f"({(user_bias['is_1star_bias'].mean()*100):.2f}%)")
print(f"ผู้ใช้ที่มีแนวโน้ม 5-star bias: {user_bias['is_5star_bias'].sum()} "
      f"({(user_bias['is_5star_bias'].mean()*100):.2f}%)")
print(f"ผู้ใช้ที่ไม่มี bias: {bias_summary['No bias']} "
      f"({bias_summary['No bias']/len(user_bias)*100:.2f}%)")
print("\nค่าเฉลี่ยของส่วนเบี่ยงเบนมาตรฐานของคะแนนผู้ใช้ทั้งหมด:",
      round(user_bias['std_rating'].mean(), 3))
````

````python
# สร้าง features ระดับ movie
def build_movie_features_from_ratings(df):
    """
    รับ dataframe แถว-level rating แล้วคืน dataframe ระดับ movie:
    - avg_rating
    - rating_count
    - rating_std
    - first_rating_date, last_rating_date (optional)
    - year (ถ้ามี)
    - title, genres
    """
    if 'datetime' in df.columns:
        df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
    agg = df.groupby(['movieid', 'title', 'genres', 'year'], as_index=False).agg(
        avg_rating = ('rating', 'mean'),
        rating_count = ('rating', 'count'),
        rating_std = ('rating', 'std'),
        first_rating = ('datetime', 'min'),
        last_rating = ('datetime', 'max'),
        is_oscar_nominated = ('is_oscar_nominated', 'max')
    )
    agg['rating_std'] = agg['rating_std'].fillna(0)
    if 'first_rating' in agg.columns and 'last_rating' in agg.columns:
        agg['rating_span_days'] = (agg['last_rating'] - agg['first_rating']).dt.days.fillna(0)
    return agg

# Preprocessing for genres (multi-label)
def split_genres_and_binarize(df, genres_col='genres', sep='|'):
    """
    รับ dataframe ระดับ movie ที่มีคอลัม genres แบบ 'A|B|C'
    คืน df ที่มีคอลัมน์ genre_... เป็น 0/1 แบบ multi-hot
    """
    df = df.copy()
    df[genres_col] = df[genres_col].fillna('').astype(str).str.strip()
    df[genres_col] = df[genres_col].apply(lambda x: x if x not in ['', '|'] else np.nan)
    df = df.dropna(subset=[genres_col])
    df['genres_list'] = df[genres_col].apply(
        lambda x: [g.strip() for g in x.split(sep) if g.strip() != ''])

    mlb = MultiLabelBinarizer()
    genres_encoded = mlb.fit_transform(df[genres_col].str.split(sep))
    genre_df = pd.DataFrame(genres_encoded, columns=[f'genre_{g}' for g in mlb.classes_])

    df = pd.concat([df.reset_index(drop=True), genre_df.reset_index(drop=True)], axis=1)
    return df, mlb

# เตรียม dataset สำหรับ modeling
def prepare_dataset_for_model(movie_df, target='is_oscar_nominated'):
    df = movie_df.copy()
    num_features = ['avg_rating', 'rating_count', 'rating_std', 'rating_span_days']

    genre_cols = [c for c in df.columns if c.startswith('genre_')]
    features = num_features + genre_cols
    X = df[features].fillna(0)
    y = df[target].astype(int)

    return X, y, num_features, genre_cols

# Modeling pipeline
def train_and_evaluate(X_train, X_test, y_train, y_test, model, model_name, use_smote=False):
    """
    ฟังก์ชันที่ยืดหยุ่นขึ้น รับ model และ model_name เข้ามาได้โดยตรง
    """
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    numeric_cols = [c for c in X_train.columns if c.startswith(('avg_rating','rating','rating_span_days'))]
    genre_cols = [c for c in X_train.columns if c.startswith('genre_')]

    preprocessor = ColumnTransformer(transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('genres', 'passthrough', genre_cols)
    ])
    if use_smote:
        pipeline = ImbPipeline(steps=[
            ('pre', preprocessor),
            ('smote', SMOTE(random_state=42)),
            ('clf', model)
        ])
    else:
        pipeline = Pipeline(steps=[
            ('pre', preprocessor),
            ('clf', model)
        ])

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    y_proba = pipeline.predict_proba(X_test)[:,1] if hasattr(model, "predict_proba") else None

    print(f"============== [{model_name}] Results ==============")
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1 Score: {f1:.4f}")
    if roc_auc is not None:
        print(f"ROC AUC: {roc_auc:.4f}")
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))

    if y_proba is not None and roc_auc is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.figure(figsize=(6,4))
        plt.plot(fpr, tpr, label=f'{model_name} ROC AUC = {roc_auc:.3f}')
        plt.plot([0,1],[0,1],'--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'{model_name} ROC Curve')
        plt.legend()
        plt.show()

    return pipeline
````

````python
# Feature Engineering
def build_movie_features_from_ratings(df):
    """
    รับ dataframe แถว-level rating แล้วคืน dataframe ระดับ movie:
    - avg_rating
    - rating_count
    - rating_std
    - first_rating_date, last_rating_date (optional)
    - title, genres
    """
    if 'datetime' in df.columns:
        df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')

    agg = df.groupby(['movieid', 'title', 'genres'], as_index=False).agg(
        avg_rating=('rating', 'mean'),
        rating_count=('rating', 'count'),
        rating_std=('rating', 'std'),
        first_rating=('datetime', 'min'),
        last_rating=('datetime', 'max'),
        is_oscar_nominated=('is_oscar_nominated', 'max')
    )

    agg['rating_std'] = agg['rating_std'].fillna(0)

    if 'first_rating' in agg.columns and 'last_rating' in agg.columns:
        agg['rating_span_days'] = (agg['last_rating'] - agg['first_rating']).dt.days.fillna(0)

    return agg

def prepare_dataset_for_model(df, target=None):
    """
    แยก features (X) และ target (y) สำหรับโมเดล
    ถ้าไม่กำหนด target จะคืนแค่ X เท่านั้น
    """
    df = df.copy()

    if target is not None and target in df.columns:
        X = df.drop(columns=[target])
        y = df[target]
    else:
        X = df
        y = None
    return X, y, None, None

#1: เตรียมข้อมูลหนังปี 2023 (Feature Engineering)
movie_df_test = build_movie_features_from_ratings(test_model.copy())

genres_dummies_test = movie_df_test['genres'].str.get_dummies(sep='|')
genres_dummies_test.columns = ['genre_' + col for col in genres_dummies_test.columns]
movie_df_test = pd.concat([movie_df_test, genres_dummies_test], axis=1)

columns_to_drop = ['first_rating', 'last_rating', 'genres', 'is_oscar_nominated', 'movieid', 'year']
movie_df_test_clean = movie_df_test.drop(columns=columns_to_drop, errors='ignore')

X_new_data_raw, _, _, _ = prepare_dataset_for_model(movie_df_test_clean, target=None)

pre = pipeline.named_steps['pre']
numeric_cols_trained = pre.transformers_[0][2]
genre_cols_trained = pre.transformers_[1][2]
X_train_cols = list(numeric_cols_trained) + list(genre_cols_trained)

X_new_data = X_new_data_raw.copy()
missing_cols = set(X_train_cols) - set(X_new_data.columns)
extra_cols = set(X_new_data.columns) - set(X_train_cols)

for col in missing_cols:
    X_new_data[col] = 0
X_new_data = X_new_data.drop(columns=extra_cols, errors='ignore')
X_new_data = X_new_data[X_train_cols]

# #2: ทำนายผล
new_predictions = pipeline.predict(X_new_data)

if hasattr(pipeline[-1], "predict_proba"):
    new_proba = pipeline.predict_proba(X_new_data)[:, 1]
else:
    new_proba = np.full(len(X_new_data), np.nan)

#3: สรุปและแสดงผลการทำนาย (20 อันดับแรก)
expected_len = len(movie_df_test)
if len(new_predictions) > expected_len:
    print(f"⚠️ ตัดผลการทำนาย (เดิม {len(new_predictions)}) ให้เหลือ {expected_len} แถว")
    new_predictions = new_predictions[:expected_len]
    new_proba = new_proba[:expected_len]

prediction_results = pd.DataFrame({
    'Title': movie_df_test['title'].values,
    'Predicted_Nomination': new_predictions,
    'Probability_of_Nomination': new_proba
})

prediction_results['Predicted_Nomination_Label'] = prediction_results['Predicted_Nomination'].apply(
    lambda x: 'Nominated (1)' if x == 1 else 'Not Nominated (0)'
)

top_20_predictions = prediction_results[['Title', 'Predicted_Nomination_Label', 'Probability_of_Nomination']].sort_values(
    by='Probability_of_Nomination', ascending=False
).head(20)

print("\n⭐ ผลการทำนายหนังปี 2023 สำหรับรางวัลปี 2024 (20 อันดับแรก):")
print("-----------------------------------------------------------------------------------------------------------------------")
print(top_20_predictions.to_string(index=False))
print("-----------------------------------------------------------------------------------------------------------------------")
````
